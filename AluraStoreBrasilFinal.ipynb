### Importação dos dados



```python
%reset -f

# Reimportar as bibliotecas certas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

##Faturamento por Loja

```python
totalLoja = loja['Preço'].sum()
totalLoja2 = loja2['Preço'].sum()
totalLoja3 = loja3['Preço'].sum()
totalLoja4 = loja4['Preço'].sum()
print(f"Loja 01 R${totalLoja:,.2f}")
print(f"Loja 02 R${totalLoja2:,.2f}")
print(f"Loja 03 R${totalLoja3:,.2f}")
print(f"Loja 04 R${totalLoja4:,.2f}")
```

# 2. Vendas por Categoria

## Vendas por categoria da Loja 01

```python
total_por_produto = loja.groupby("Produto")["Preço"].sum().reset_index().sort_values(by="Preço", ascending=False)
total_por_produto["Preço"] = total_por_produto["Preço"].apply(lambda x: f"R${x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
total_por_produto.head(10)
```

# 3. Média de Avaliação das Lojas


```python
media_avaliacao = loja['Avaliação da compra'].mean()
media_avaliacao2 = loja2['Avaliação da compra'].mean()
media_avaliacao3 = loja3['Avaliação da compra'].mean()
media_avaliacao4 = loja4['Avaliação da compra'].mean()

print(f"Média de avaliação da loja 01: {media_avaliacao:.2f}")
print(f"Média de avaliação da loja 02: {media_avaliacao2:.2f}")
print(f"Média de avaliação da loja 03: {media_avaliacao3:.2f}")
print(f"Média de avaliação da loja 04: {media_avaliacao4:.2f}")

medias = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Média Avaliação': [
        loja['Avaliação da compra'].mean(),
        loja2['Avaliação da compra'].mean(),
        loja3['Avaliação da compra'].mean(),
        loja4['Avaliação da compra'].mean()
    ]
})
plt.figure(figsize=(8,5))
plt.plot(
    medias['Loja'],
    medias['Média Avaliação'],
    marker='o',
    linestyle='-',
    color='royalblue',
    linewidth=2
)


plt.title('Média de Avaliação das Compras por Loja', fontsize=14)
plt.xlabel('Loja', fontsize=12)
plt.ylabel('Média da Avaliação', fontsize=12)
plt.ylim(0, 5)  # se as avaliações forem de 0 a 5
plt.yticks(np.arange(0, 5.1, 0.30))
plt.grid(True, linestyle='--', alpha=0.6)  # grade leve
plt.tight_layout()
plt.show()

```

# 4. Produtos Mais e Menos Vendidos

```python

top10_produtos = (
    loja["Produto"]
    .value_counts()
    .reset_index()
    .rename(columns={"index": "Produto", "count": "quantidade_vendida"})
    .head(10)
)


print("Loja 1 Top 10 produtos mais vendidos:")
print(top10_produtos)

plt.figure(figsize=(10,5))
plt.bar(top10_produtos["Produto"], top10_produtos["quantidade_vendida"], color='skyblue')
plt.title("Top 10 Produtos Mais Vendidos - Loja 1")
plt.xlabel("Produto")
plt.ylabel("Quantidade Vendida")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja

```python
frete_medio = loja['Frete'].mean()
frete_medio_L2 = loja2['Frete'].mean()
frete_medio_L3 = loja2['Frete'].mean()
frete_medio_L4 = loja2['Frete'].mean()

print(f"Frete médio da loja 01: R${frete_medio:.2f}")
print(f"Frete médio da loja 02: R${frete_medio_L2:.2f}")
print(f"Frete médio da loja 03: R${frete_medio_L3:.2f}")
print(f"Frete médio da loja 04: R${frete_medio_L4:.2f}")
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
valores = [frete_medio, frete_medio_L2, frete_medio_L3, frete_medio_L4]

plt.figure(figsize=(7,7))
plt.pie(
    valores,
    labels=lojas,
    autopct='%1.1f%%',     # mostra porcentagem com 1 casa decimal
    startangle=90,         # começa o gráfico no topo
    colors=['skyblue', 'lightgreen', 'salmon', 'orange'],
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Proporção do Frete Médio por Loja')
plt.show()
```
